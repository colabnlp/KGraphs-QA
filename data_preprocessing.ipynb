```python
import re
import string
import nltk
from nltk.corpus import stopwords
```

```python
path = "/Users/rahulsm/Downloads/concept_assertion_relation_training_data/beth/txt/record-13.txt"
path2 = "/Users/rahulsm/Downloads/concept_assertion_relation_training_data/beth/concept/record-13.con"
```

```python
new_data = []
with open(path, 'r') as data:
    for lines in data.readlines():
        for words in lines.split():
            words = words.lower()
            if re.match("^[a-z0-9_]*$", words):
                new_data.append(words)
```

```python
ner_dict = dict()
with open(path2, 'r') as ner:
    for line in ner.readlines():
        line = line.lower()
        matches=re.findall(r'\"(.+?)\"',line)
        ner_dict[matches[0]] = matches[1]
print(len(new_data))
print(new_data)
```

```python
def all_occurences(file, str):
    initial = 0
    while True:
        initial = file.find(str, initial)
        if initial == -1: return -1
        yield initial
        initial += len(str)
```

```python
stop_words = set(stopwords.words('english'))
stop_words.update(('admission','date','discharge','birth', 'sex', 'report', 'end'))
new_data = [w for w in new_data if not w in stop_words and len(w)>2 and w.isalpha()]
# print(len(new_data))
# print(new_data)

ners = dict()
print(len(ner_dict))
for keys in ner_dict: 
    keys = keys.lower()
    words = [w for w in keys.split() if not w in stop_words and w.isalpha() and re.match("^[a-z0-9_]*$", w)]
    word = ' '.join(map(str, words))
    if len(words)>0:
        ners[word] = ner_dict[keys]
print(len(ners))
for keys in ners:
    print(f"{keys} for {ners[keys]}")
#     print(word)
#     print(keys)
```
